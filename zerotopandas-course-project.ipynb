A high correlation between two variables indicates that they are positively related, while a low correlation indicates that they are not related or negatively related. #### Q1: How did the SPY ETF perform during the 2008 Financial Crisis and Covid-19 Pandemic ?


````python
# Select data for the two crises
crisis_2008 = df.loc['2008-01-01':'2009-12-31']
crisis_covid = df.loc['2020-01-01':'2021-12-31']

# Create a line plot of the two crises
plt.plot(crisis_2008['Close'], label='2008 Financial Crisis')
plt.plot(crisis_covid['Close'], label='COVID-19 Pandemic')

# Set the title and labels for the graph
plt.title('SPY ETF Performance During Crises')
plt.xlabel('Date')
plt.ylabel('Closing Price')

# Add a legend
plt.legend()

# Show the graph
plt.show()
````

We can see that the 2008 financial crisis made the market dip initially and the effects were not recovered from even till the 2010s. As for the covid-19 pandemic we can see that the initial cliff dive of closing price of the SPY ETF was swift and sudden but its recovery was exponentially fast and even eclipsing the closing price before the pandemic happened.


#### Q2: How many trading days had a volume above average volume?

````python
# Calculate the average volume
avg_volume = df['Volume'].mean()

# Count the number of trading days with volume greater than the average
num_high_volume_days = (df['Volume'] > avg_volume).sum()

#Count number of trading days with volume lower than average
num_low_volume_days = (df['Volume'] < avg_volume).sum()

print(f"There were {num_high_volume_days} trading days with volume above the average ({avg_volume:.2f})")
print(f"However there are also {num_low_volume_days} trading days with volume below the average" , num_low_volume_days)


# Create a histogram of the volume column
plt.hist(df['Volume'], bins=50)
plt.axvline(avg_volume, color='red', linestyle='dashed', linewidth=2, label='Average Volume')
plt.xlabel('Volume')
plt.ylabel('Count')
plt.legend()
plt.show()
````

We can that the red denotes the average volume. At a glance we can see that the count of lower than average trading volume days are more numerous compared to days with higher than average trading volume. There we can conclude that over the past 30years , more often that not volume is typically low and only goes above average volume if there is significant events happening in the world economy etc

#### Q4: What is the overall trend in SPY ETF price over time?

````python
# Compute the rolling mean of the close price over a 30-day window
rolling_mean = df['Close'].rolling(window=30).mean()

# Create a line plot of the close price and its rolling mean over time
fig, ax = plt.subplots()
ax.plot(df.index, df['Close'], label='Close Price')
ax.plot(df.index, rolling_mean, label='30-Day Rolling Mean')
ax.set_title('SPY ETF Price Over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.legend()

# Show the graph
plt.show()
````

Y axis represents SPY ETF price, X axis represents time period.
We can see that the close price and the 30d rolling mean are almost similar to one another other than the fact that 30d rolling mean has higher high and lower low compared to close price. This result is what we should expect as 30d rolling mean is meant to be a smoothed version of the close price over time due to the way it is calculated by taking the average of the close price over the past 30days. 30d rolling mean should provide a more stable long term overall view of the trend itself.

#### Q5: TODO - Are there any pattern or trends in the volatility of the SPY ETF ?

````python
# Compute the rolling standard deviation of the close price over a 30-day window
rolling_std = df['Close'].rolling(window=30).std()

# Create a line plot of the rolling standard deviation of the close price over time
fig, ax = plt.subplots()
ax.plot(df.index, rolling_std)
ax.set_title('SPY ETF Volatility Over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Volatility')

# Show the graph
plt.show()
````

Standard deviation is used to quantify the volatility of SPY ETF in this scenario. As seen in the graph above, the year 2008 and year 2020 shows a significant high volatility compared to the other years. This also corresponds accordingly with the 2008 financial crisis and the covid 19 pandemic period. I can see that during times of significant events that affect the world economy. SPY volatility is also correspondingly high during those periods.
